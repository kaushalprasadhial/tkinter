```python
#just run the file
import tkinter as tk
from PIL import Image, ImageTk
import requests
from tkinter import font

def fer_to_cel(Fahrenheit):
    return (Fahrenheit - 32) * 5.0/9.0

def format_result(weather):
    degree_sign = u'\N{DEGREE SIGN}'
    city_name = weather['name']
    country_name = weather['sys']['country']
    temp = fer_to_cel(weather['main']['temp'])
    pressure = weather['main']['pressure']
    humidity = weather['main']['humidity']
    temp_min = fer_to_cel(weather['main']['temp_min'])
    temp_max = fer_to_cel(weather['main']['temp_max'])
    main_weather = weather['weather'][0]['main']
    discription = weather['weather'][0]['description']
    first_line = 'City: %s, Country: %s' %(city_name, country_name)
    second_line = 'Temprature: %.2s, Min. temprature: %.2s, Max. temprature: %.2s in %sC' %(temp, temp_min, temp_max, degree_sign)
    third_line = 'Average weather: %s, Discription: %s' %(main_weather, discription)
    
    return first_line + '\n' + second_line + '\n' + third_line

#function to get the weather
def get_weather(location):
    api_key = '857e5786894a74174ef4083c2731fa54'
    api_call_url = 'https://api.openweathermap.org/data/2.5/weather'
    #api_call_url = 'https://api.openweathermap.org/data/2.5/forecast'
    params = {'APPID':api_key,
             'q':location,
             'units':'imperial'}
    request = requests.get(api_call_url, params=params)
    weather = request.json()
    print(weather)
    label['text'] = format_result(weather)
    
root = tk.Tk()

Image = Image.open("Unique_Wallpaper_36.jpg")
#backgroundimage = ImageTk.PhotoImage(image)
backgroundimage = tk.PhotoImage(file='Screenshot (15).png')

canvas= tk.Canvas(root, height=400,width=900, bg='#3399FF')
canvas.pack()

background_label = tk.Label(root, image=backgroundimage)
background_label.image = backgroundimage
background_label.place(relwidth=1,relheight=1)


frame = tk.Frame(root,bg='#0066FF', bd=5)
frame.place(relx=0.5, rely=0.1,relwidth=0.75,relheight=0.2,anchor='n')

entry = tk.Entry(frame,bg='#3399FF',font=('Candara',20))
entry.place(relwidth=0.65, relheight=1)

button = tk.Button(frame, text = 'get weather', command=lambda: get_weather(entry.get()), font=('Microsoft JhengHei Light',10))
button.place(relx=0.7,relheight=1,relwidth=0.3)

lower_frame = tk.Frame(root,bg='#0066FF', bd=5)
lower_frame.place(relx=0.5, rely=0.5,relwidth=0.75,relheight=0.4,anchor='n')

label = tk.Label(lower_frame, text='Enter the name of the city in above text box\n and click the button "get weather"', font=('Candara',10))
label.place(relwidth=1,relheight=1)

root.mainloop()
```
